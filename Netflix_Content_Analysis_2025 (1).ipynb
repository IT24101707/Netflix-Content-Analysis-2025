```python
# ---------------- MOVIES DATA CLEANING ----------------

# Remove duration column (all values are missing)
movies.drop(columns=['duration'], inplace=True)

# Fill missing values
movies['director'] = movies['director'].fillna('Unknown')
movies['cast'] = movies['cast'].fillna('Unknown')
movies['country'] = movies['country'].fillna('Unknown')
movies['genres'] = movies['genres'].fillna('Unknown')
movies['description'] = movies['description'].fillna('No Description')

print("Movies dataset cleaned successfully!")
```

```python
# ---------------- TV SHOWS DATA CLEANING ----------------

tv['director'] = tv['director'].fillna('Unknown')
tv['cast'] = tv['cast'].fillna('Unknown')
tv['country'] = tv['country'].fillna('Unknown')
tv['genres'] = tv['genres'].fillna('Unknown')
tv['description'] = tv['description'].fillna('No Description')

print("TV Shows dataset cleaned successfully!")
```

```python
print(movies.isnull().sum())
print("\n----------------------\n")
print(tv.isnull().sum())
```

```python
plt.figure(figsize=(12,6))

movies['release_year'].value_counts().sort_index().plot()

plt.title("Netflix Movies Released by Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")

plt.show()
```

```python
plt.figure(figsize=(12,6))

movies['genres'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Movie Genres")
plt.xlabel("Genre")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

movies['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Movies")
plt.xlabel("Country")
plt.ylabel("Movies")

plt.xticks(rotation=45)

plt.show()
```

```python
top_rated = movies.sort_values(by='vote_average', ascending=False)

top10 = top_rated[['title','vote_average']].head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='vote_average',
    y='title'
)

plt.title("Top 10 Highest Rated Netflix Movies")
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")

plt.show()
```

```python
popular = movies.sort_values(by='popularity', ascending=False)

top10 = popular[['title','popularity']].head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='popularity',
    y='title'
)

plt.title("Top 10 Most Popular Netflix Movies")
plt.xlabel("Popularity Score")
plt.ylabel("Movie Title")

plt.show()
```

```python
plt.figure(figsize=(12,6))

movies['language'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Languages")
plt.xlabel("Language")
plt.ylabel("Number of Movies")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    movies['vote_average'],
    bins=20,
    kde=True
)

plt.title("Distribution of Movie Ratings")
plt.xlabel("Vote Average")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = movies[['popularity',
               'vote_average',
               'vote_count',
               'budget',
               'revenue']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
top_revenue = movies.sort_values(by='revenue', ascending=False)

top10 = top_revenue[['title','revenue']].head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='revenue',
    y='title'
)

plt.title("Top 10 Highest Revenue Movies")
plt.xlabel("Revenue")
plt.ylabel("Movie Title")

plt.show()
```
